Есть точечные соответвия $x_i \leftrightarrow x_i'$ получить $X_i$
Мы не можем востановить абсолютные координаты, мы можем узнать только координаты сдвига, или т.е координаты в какой то системе координат, и можно получить их путем сдвига и переноса. Так же мы не можем востановить масштаб.

$ x_i = PX \\ x_i' = P'X' $

1. Найти фундаментальную матрицу $x_i''Fx_i = 0 $, требуется 8 точечных соответсвий
2. По матрице F найти $P, P', P = [I|0], P' = [[e_x']F|e']$
3. Для каждого точечного соответсвия $x_i \leftrightarrow x_i'$ найти $X_i$, Триангуляция

$ x_i^\prime Fx_i = 0 \\ x = H_sX_i \\ P = PH_S^{-1} \\ P' = P'H_s^{-1} $

##### Теорема проективного востановления
Пусть $x_i \leftrightarrow x_i' $ это набор точечных соответствий. Причем матрица $F$ определяется единственным образом из уравнений $x_i^T = 0$ пусть $ (P_1, P_1', ( x_{1i} )) $ и $ (P_2, P_2', ( x_{2i} )) $, тогда существует невырожденная матрица $H$, такая что $ P_2 = P_1 H^{-1}, P_2'= P_1' H^{-1}, X_{2i} = HX_{1i} $ для всех $i$, за исключением тех $i$ для которых $ F_{x_i} = x_i'TF=0 $
